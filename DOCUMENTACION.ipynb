## Documentación del Código

1. **Importación del módulo Tkinter**

   Se importa el módulo `tkinter` para crear interfaces gráficas en Python, asignándole el alias `tk` para simplificar el código.

2. **Definición de la clase `Application`**

   La clase `Application` hereda de `tk.Frame`, y representa la ventana principal de la aplicación.

   2.1 **Constructor de la clase**

   El constructor inicializa la ventana principal (`master`) y crea los widgets llamando al método `create_widgets`. Además, se definen dos diccionarios vacíos: `datos_persona` y `claves_especificas`, que se usarán para almacenar datos ingresados por el usuario.

   2.2 **Método `create_widgets`**

   Este método se encarga de crear y posicionar todos los widgets de la interfaz gráfica, incluyendo etiquetas, entradas de texto y botones.

   - **Entrada y etiqueta para el nombre**

     Se crean una etiqueta y una entrada de texto para que el usuario ingrese su nombre. Ambos widgets se colocan en la parte superior de la ventana utilizando el método `pack`.

   - **Entrada y etiqueta para la edad**

     De manera similar, se crean una etiqueta y una entrada para la edad del usuario.

   - **Entrada y etiqueta para la ciudad**

     Se crean una etiqueta y una entrada de texto para ingresar la ciudad.

   - **Entrada y etiqueta para la clave específica**

     Se crea otra etiqueta y entrada donde el usuario puede ingresar una clave específica que servirá para almacenar y acceder a los datos.

   - **Botón para agregar datos**

     Se crea un botón llamado "Agregar datos", que ejecuta el método `agregar_datos` cuando es presionado. Este método se encarga de agregar la información ingresada por el usuario al diccionario `claves_especificas`.

   - **Botón para mostrar valor de una clave específica**

     Este botón ejecuta el método `mostrar_valor`, que mostrará la información almacenada bajo una clave específica.

   - **Etiqueta para mostrar resultados**

     Se crea una etiqueta vacía donde se mostrarán los resultados de la búsqueda de claves específicas.

   - **Botón para salir**

     Este botón cierra la aplicación cuando es presionado al ejecutar `self.master.destroy`.

3. **Método `agregar_datos`**

   Este método agrega los datos ingresados a un diccionario llamado `claves_especificas`. Se obtiene el valor de la clave específica ingresada por el usuario y se convierte a minúsculas. Si la clave no existe en el diccionario, se crea una nueva entrada. Los datos de nombre, edad y ciudad se almacenan bajo esa clave, y se imprime el contenido del diccionario en la consola.

4. **Método `mostrar_valor`**

   Este método busca la clave específica ingresada y muestra su valor en la etiqueta `resultado_label`. Convierte la clave ingresada a minúsculas y si la clave existe en el diccionario `claves_especificas`, se muestra su valor. Si la clave no existe, se muestra un mensaje indicando que no se encontró.

5. **Ejecución de la aplicación**

   Finalmente, se crea la ventana principal utilizando `tk.Tk()`, se inicializa la aplicación con `Application(master=root)` y se llama al bucle principal de Tkinter con `app.mainloop()` para que la ventana permanezca abierta y reactiva.
